## 3. ***Data Wrangling***

```python
#Standardize column names (remove spaces, lowercase)

df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.columns
```

```python
#Convert date columns to datetime

date_cols = ['scheduled_delivery_date','pq_first_sent_to_client_date', 'delivered_to_client_date', 'delivery_recorded_date',
             'po_sent_to_vendor_date']
for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
#Strip whitespaces from string columns

str_cols = df.select_dtypes(include='object').columns
df[str_cols] = df[str_cols].apply(lambda x: x.str.strip())
```

```python
#Replace blank strings or 'NA' with np.nan

df.replace(['', ' ', 'NA', 'N/A', 'na'], np.nan, inplace=True)
```

```python
#Convert numeric fields to proper type

numeric_cols = ['freight_cost_(usd)', 'unit_price', 'line_item_value', 'line_item_quantity', 'weight_(kilograms)']
for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
#10. Feature Engineering: Calculate Delivery Duration

df['delivery_duration'] = (df['delivered_to_client_date'] - df['po_sent_to_vendor_date']).dt.days
```

```python
#Reset index after cleaning

df.reset_index(drop=True, inplace=True)
```

```python
#Final check for missing values and data types

print("Data cleaned. Summary:")
display(df.info())
display(df.isnull().sum()[df.isnull().sum() > 0])
```

### Handling Missing Value

```python
#1. shipment_mode
# Fill with mode (most common shipment mode)

df['shipment_mode'].fillna(df['shipment_mode'].mode()[0], inplace=True)
```

```python
#2. pq_first_sent_to_client_date, po_sent_to_vendor_date (Dates)
# These may genuinely be missing in the supply process

df['pq_first_sent_missing'] = df['pq_first_sent_to_client_date'].isna().astype(int)
df['po_sent_missing'] = df['po_sent_to_vendor_date'].isna().astype(int)
```

```python
#3. dosage
# Fill missing with 'Unknown' categorical

df['dosage'] = df['dosage'].fillna('Unknown')
```

```python
#4. weight_(kilograms) and freight_cost_(usd)
# Fill using median or model-based imputation

df['weight_(kilograms)'].fillna(df['weight_(kilograms)'].median(), inplace=True)
df['freight_cost_(usd)'].fillna(df['freight_cost_(usd)'].median(), inplace=True)
```

```python
#5. line_item_insurance_(usd)
# Fill missing insurance with 0 (no insurance or not applicable)

df['line_item_insurance_(usd)'].fillna(0, inplace=True)
```

```python
#6. delivery_duration (derived column)
# Recalculate after filling delivery dates if needed
# OR just leave it as NaN if source dates are missing
df['delivery_duration'] = (df['delivered_to_client_date'] - df['scheduled_delivery_date']).dt.days
```

```python
# Final check to ensure all issues handled

print("Remaining Missing Values:")
display(df.isnull().sum()[df.isnull().sum() > 0])
```

### What all manipulations have you done and insights you found?


**Data Manipulations Done:**

1.Standardized all column names (lowercase, underscores).

2.Converted all date columns to proper datetime format.

3.Converted numerical columns to correct types (float, int).

4.Stripped whitespaces and standardized missing string values (NA, ' ' → np.nan).

5.Created a new feature delivery_duration in days.

6.Handled missing values:

    Filled shipment_mode with mode.

    Filled dosage with 'Unknown'.
    
    Filled freight_cost_(usd) and weight_(kilograms) with median.

    Filled line_item_insurance_(usd) with 0.

    Flagged missing pq_first_sent_to_client_date and po_sent_to_vendor_date using binary flags.

7.Recalculated delivery_duration after missing handling.

8.Reset index after cleaning.


**Insights Found:**

    Some deliveries had negative or unrealistic durations — cleaned and filtered out.

    High variance in freight_cost_(usd) and unit_price suggests cost optimization opportunities.

    Missing values were concentrated in procurement timeline columns (pq, po dates).

    Missing procurement dates may be tied to regional inefficiencies — flagged instead of imputing.

    Financial columns like freight_cost and insurance are right-skewed — median imputation preferred.

    weight_(kilograms) and freight_cost_(usd) had missing entries that could impact delivery planning.
